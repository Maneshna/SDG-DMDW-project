```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/sustainable_development_report_2023.csv")
df.head()
```

```python
# === Sustainable Development Dataset: Cleaning + Visualization + Linear Regression ===

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error



print("Missing values:\n", df.isnull().sum())


df_mean = df.copy()
df_mean.fillna(df_mean.mean(numeric_only=True), inplace=True)


df_median = df.copy()
df_median.fillna(df_median.median(numeric_only=True), inplace=True)


df_mode = df.copy()
for col in df_mode.select_dtypes(include=[np.number]).columns:
    mode_val = df_mode[col].mode()[0]
    df_mode[col].fillna(mode_val, inplace=True)


df_clean = df_median

print("\nMissing values after cleaning:\n", df_clean.isnull().sum())


# 1) Histogram
df_clean.hist(figsize=(15,12), bins=20)
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()

# 2) Bar chart:
plt.figure(figsize=(8,5))
df_clean['region'].value_counts().plot(kind='bar')
plt.title("Countries per Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()

# 3) Scatter plots
for col in ['goal_1_score', 'goal_10_score', 'goal_14_score']:
    plt.figure(figsize=(6,4))
    plt.scatter(df_clean[col], df_clean['overall_score'], alpha=0.7)
    plt.title(f"Overall Score vs {col}")
    plt.xlabel(col)
    plt.ylabel("Overall Score")
    plt.show()

# 4) Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df_clean.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()

```
