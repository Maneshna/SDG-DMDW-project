```python
# Colab-ready end-to-end pipeline for data cleaning, visualization & model training
# Adjust FILE_NAME if your CSV filename differs.


# ---- Imports ----
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, r2_score, mean_squared_error
import joblib
from google.colab import files



# ---- Quick overview ----
print("\nColumn dtypes:")
print(df.dtypes)
print("\nMissing values (count):")
print(df.isnull().sum())

# ---- Choose target column ----
# Preference: 'target','label','y' else last column
preferred_targets = ['target', 'label', 'y']
target_col = None
for t in preferred_targets:
    if t in df.columns:
        target_col = t
        break
if target_col is None:
    target_col = df.columns[-1]
print(f"\nUsing target column: '{target_col}'")

# ---- Separate features/target ----
X = df.drop(columns=[target_col])
y = df[target_col]

# ---- Identify numeric and categorical columns ----
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()
print("\nNumeric columns:", numeric_cols)
print("Categorical columns:", cat_cols)

# ---- Imputation strategy ----
# Options for numeric_impute: 'auto' | 'mean' | 'median' | 'mode'
numeric_impute = 'auto'   # change to 'mean' or 'median' or 'mode' if you prefer explicit

X_imputed = X.copy()

# Numeric imputation
for col in numeric_cols:
    if X_imputed[col].isnull().any():
        if numeric_impute == 'mean':
            val = X_imputed[col].mean()
            X_imputed[col].fillna(val, inplace=True)
            print(f"Filled numeric column '{col}' with MEAN = {val:.4f}")
        elif numeric_impute == 'median':
            val = X_imputed[col].median()
            X_imputed[col].fillna(val, inplace=True)
            print(f"Filled numeric column '{col}' with MEDIAN = {val:.4f}")
        elif numeric_impute == 'mode':
            mode_vals = X_imputed[col].mode(dropna=True)
            if len(mode_vals) > 0:
                val = mode_vals[0]
            else:
                val = 0
            X_imputed[col].fillna(val, inplace=True)
            print(f"Filled numeric column '{col}' with MODE = {val}")
        else:  # auto: use median if strongly skewed, else mean
            skew = X_imputed[col].dropna().skew()
            if abs(skew) > 1:
                val = X_imputed[col].median()
                X_imputed[col].fillna(val, inplace=True)
                print(f"Auto-filled numeric column '{col}' with MEDIAN (skew={skew:.2f}) = {val:.4f}")
            else:
                val = X_imputed[col].mean()
                X_imputed[col].fillna(val, inplace=True)
                print(f"Auto-filled numeric column '{col}' with MEAN (skew={skew:.2f}) = {val:.4f}")

# Categorical imputation: mode (most frequent)
for col in cat_cols:
    if X_imputed[col].isnull().any():
        modes = X_imputed[col].mode(dropna=True)
        if len(modes) > 0:
            val = modes[0]
        else:
            val = "MISSING"
        X_imputed[col].fillna(val, inplace=True)
        print(f"Filled categorical column '{col}' with MODE = {val}")

# Target missing handling: drop rows with missing target
if y.isnull().sum() > 0:
    print(f"Target '{target_col}' has {y.isnull().sum()} missing rows - dropping them.")
    mask = y.notnull()
    X_imputed = X_imputed[mask]
    y = y[mask]

print("\nMissing values after imputation (should be zero):")
print(X_imputed.isnull().sum()[X_imputed.isnull().sum() > 0])

# ---- Encoding categorical features ----
X_encoded = pd.get_dummies(X_imputed, columns=cat_cols, drop_first=True)
print("\nShape after one-hot encoding:", X_encoded.shape)

# ---- Scaling numeric features ----
scaler = StandardScaler()
num_after = X_encoded.select_dtypes(include=[np.number]).columns.tolist()
X_encoded[num_after] = scaler.fit_transform(X_encoded[num_after])
print("Scaled numeric features using StandardScaler.")

# ---- Visualizations ----
# Histograms (numeric)
print("\nPlotting histograms for numeric columns...")
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.hist(X_imputed[col].dropna(), bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Bar charts for categorical columns (top counts)
print("\nPlotting bar charts for categorical columns (top 10 categories)...")
for col in cat_cols:
    plt.figure(figsize=(8,4))
    vc = X_imputed[col].value_counts().nlargest(10)
    plt.bar(vc.index.astype(str), vc.values)
    plt.title(f'Top categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Scatter plots: up to 3 numeric columns vs target (if target numeric) or vs first numeric if categorical target
print("\nPlotting scatter plots (up to 3 numeric features) vs target...")
numeric_for_scatter = numeric_cols[:3]
is_target_numeric = pd.api.types.is_numeric_dtype(y)
for col in numeric_for_scatter:
    plt.figure(figsize=(6,4))
    plt.scatter(X_imputed[col], y, alpha=0.6, s=12)
    plt.title(f'{col} vs {target_col}')
    plt.xlabel(col)
    plt.ylabel(target_col)
    plt.tight_layout()
    plt.show()

# ---- Train/test split ----
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
print("\nTrain/test sizes:", X_train.shape, X_test.shape)

# ---- Model selection & training ----
model_artifacts = {}
if is_target_numeric:
    print("\nTarget is numeric -> training RandomForestRegressor")
    from sklearn.ensemble import RandomForestRegressor
    model = RandomForestRegressor(n_estimators=200, random_state=42)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    r2 = r2_score(y_test, preds)
    mse = mean_squared_error(y_test, preds)
    print(f"R^2 on test set: {r2:.4f}")
    print(f"MSE on test set: {mse:.4f}")
    model_artifacts['target_type'] = 'regression'
else:
    print("\nTarget is categorical -> training RandomForestClassifier")
    # encode target labels consistently
    unique_vals = pd.Series(pd.concat([y_train, y_test]).unique())
    mapping = {v: i for i, v in enumerate(unique_vals)}
    y_train_enc = y_train.map(mapping)
    y_test_enc = y_test.map(mapping)
    model = RandomForestClassifier(n_estimators=200, random_state=42)
    model.fit(X_train, y_train_enc)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test_enc, preds)
    print(f"Accuracy on test set: {acc:.4f}")
    print("\nClassification report:")
    print(classification_report(y_test_enc, preds))
    print("Confusion matrix:")
    print(confusion_matrix(y_test_enc, preds))
    model_artifacts['target_type'] = 'classification'
    model_artifacts['target_mapping'] = mapping

# ---- Save model + preprocessing objects to Drive ----
OUT_MODEL_PATH = f"/content/drive/MyDrive/Colab Notebooks/trained_model_pipeline.pkl"
joblib.dump({
    'model': model,
    'scaler': scaler,
    'feature_columns': X_encoded.columns.tolist(),
    'target_col': target_col,
    'artifacts': model_artifacts
}, OUT_MODEL_PATH)
print(f"\nSaved model + preprocessing to: {OUT_MODEL_PATH}")

# Optional: provide download link to local machine
try:
    files.download(OUT_MODEL_PATH)
except Exception as e:
    print("files.download() may not work in all browsers/environments. Model is saved to Drive at:", "/content/drive/MyDrive/Colab Notebooks/trained_model_pipeline.pkl")

# ---- Summary ----
summary = {
    'dataset_shape': df.shape,
    'numeric_columns': numeric_cols,
    'categorical_columns': cat_cols,
    'target_column': target_col,
    'model_type': 'RandomForestRegressor' if is_target_numeric else 'RandomForestClassifier'
}
print("\nSummary:")
print(summary)
```
